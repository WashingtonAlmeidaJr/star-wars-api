```python
#importando bibliotecas

import requests
import json
import pandas as pd
from pandas import json_normalize
```

```python

#tabela concatenada
tabela_final=pd.DataFrame()
pagina=1


#loop para passar pelas paginas da api de naves
while pagina in range(5):
    #fazer requisio da api 
    link=f'https://swapi.dev/api/starships/?page={pagina}'
    request=requests.get(link)
    request_json=request.json()
    tabela=pd.DataFrame(request_json['results'])
    tabela_final=pd.concat([tabela_final,tabela])
    pagina=pagina+1
tabela_final

```
